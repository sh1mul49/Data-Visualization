```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create lsit x & y and then plot a dotted plot
x=[1,2,3,4,5]
y=[50,60,52,55,62]
plt.plot(x,y,color="red",linewidth='7',linestyle='dotted')
plt.title("plot between x and y")
plt.xlabel("x value")
plt.ylabel("y value")
```

```python
# Create list x & y and then plot a scatter plot
x=[1,2,3,4,5]
y=[50,60,52,55,62]
plt.scatter(x,y)
```

```python
# same like previous on but except here specify the plot color as green
x=[1,2,3,4,5]
y=[50,60,52,55,62]
plt.plot(x,y,'rd',color="green")
```

```python
# Create 3 list randomely and plot together 3 list in a same plot
days=[1,2,3,4,5,6]
max_t=[50,51,52,53,54,48]
min_t=[42,41,43,40,38,33]
avg_t=[42,41,43,40,35,34]
plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")
plt.legend("best",shadow='true')
plt.grid()
```
